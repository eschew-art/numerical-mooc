```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ms=50
g=9.81
rho=1.091
r=.5
A=(np.pi)*(r**2)
ve=325
Cd=0.15

mp0=100
t0=0
dt=.1
v_i=0
h_i=0
h_i1=1 #writing any number not equal to zero
v_i1=0
n=0
t=[]
h=[]
v=[]
mp=[]
t.append(t0)
v.append(v_i)
h.append(h_i)
mp.append(0)
```

```python
while h_i1 > -5 :
        if n>=5000:
            break

        n=n+1

        t_i=t0+n*dt
        t.append(t_i)

        mp_dot=20
        if t_i<5:
            mp.append(mp0-mp_dot*t_i)
        else :
            mp.append(mp0-mp_dot*5)
       # print(mp)

        v_i1 = v_i + (dt/(ms+mp[n]))*(-(ms+mp[n])*g + mp_dot*ve - 0.5 * rho*v_i*abs(v_i)*A*Cd)
        v.append(v_i1)

        h_i1=h_i + t_i*v_i
        h.append(h_i1)

        v_i=v_i1
        h_i=h_i1

j=t.index(3.2)
print(mp[j])
        
#plt.plot(t,h)
#plt.plot(t,v)
#print(mp)
#print(t)
#print(h)
print(v)
```
